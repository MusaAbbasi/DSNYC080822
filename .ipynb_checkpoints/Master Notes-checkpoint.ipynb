```python
def fun_estimator(input_series):
    
    X_centered = input_series - input_series.mean()
    fun_est = np.abs(X_centered).sum()
    return fun_est
```

```python
sns.barplot(x = 'species', y = 'body_mass_g', data = penguins_df, ci = 99.99, estimator = fun_estimator)
plt.show()
```

```python
sns.pairplot(penguins_df, corner = True)
plt.show()
```

```python
sns.pairplot(penguins_df, hue = 'species', 
             corner = True)
plt.show()
```

```python
sns.pairplot(penguins_df, hue = 'species', diag_kind= "hist", kind = "hist", 
             diag_kws = {'element': 'step'}, plot_kws = {'bins': 15}, corner = True)
plt.show()
```

### Probabilities

Naturally lead to Bayes' theorem: we'll interpret later.

$$ P(X = x_i | Y = y_j)P(Y = y_j) = P(Y = y_j | X = x_i)P(X = x_i)$$
<br><br>

implies:

<center><b>Bayes' Theorem</b></center>

$$ P(X = x_i | Y = y_j) = \frac{P(Y = y_j | X = x_i)P(X = x_i)}{P(Y = y_j)} $$

### Combinatorics 

In general for N objects:

$$ N! = N(N-1)(N-2)(N-3)....2\times1 $$ is the number of permutations (or arrangements)

The general permutation case: n objects, r bins.

Total arrangements: AKA Total posibilities taking order into account

$$ P(n,r) = \frac{n!}{(n-r)!} $$

General formula for combinations AKA total posibilities regardless of order
$$ C(N, r) = {N \choose r} =   \frac{N!}{(N - r)! r!} $$

- Arrangements within given unselected group: $ (N - r)! $ 
-  Arrangements within given selected group: $ r! $
- Total number of arrangements selected/unselected: $ N! $
